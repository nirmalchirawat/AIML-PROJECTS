```python
!pip install scikit-learn pandas
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.preprocessing import LabelEncoder, StandardScaler


# Dataset

data = {
    "age": [22, 25, 30, 35, 40, 45, 50, 28, 33, 38],
    "income": [20000, 25000, 40000, 50000, 60000, 70000, 80000, 45000, 52000, 58000],
    "credit_score": [600, 620, 650, 680, 700, 720, 750, 670, 690, 710],
    "employment": ["No", "Yes", "Yes", "Yes", "Yes", "Yes", "Yes", "Yes", "Yes", "Yes"],
    "loan_amount": [100000, 150000, 250000, 350000, 450000, 550000, 650000, 300000, 400000, 500000]
}

df = pd.DataFrame(data)


# Encode categorical column

encoder = LabelEncoder()
df["employment"] = encoder.fit_transform(df["employment"])


# Features & Targets

X = df[["age", "income", "credit_score", "employment"]]
y_loan = df["loan_amount"]
y_approval = (df["loan_amount"] >= 300000).astype(int)


# Feature Scaling

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# Train Models

linear_model = LinearRegression()
linear_model.fit(X_scaled, y_loan)

logistic_model = LogisticRegression(max_iter=1000)
logistic_model.fit(X_scaled, y_approval)


# User Input

print("Employment options:", list(encoder.classes_))

age = int(input("Enter age: "))
income = float(input("Enter monthly income: "))
credit_score = int(input("Enter credit score: "))
employment = input("Employed (Yes/No): ").strip().title()

if employment not in encoder.classes_:
    raise ValueError("Employment must be Yes or No")

employment_encoded = encoder.transform([employment])[0]


# Prediction

user_data = [[age, income, credit_score, employment_encoded]]
user_data_scaled = scaler.transform(user_data)

predicted_loan = linear_model.predict(user_data_scaled)
approval = logistic_model.predict(user_data_scaled)
probability = logistic_model.predict_proba(user_data_scaled)[0][1]


# Output

print("\n--- LOAN RISK ANALYSIS RESULT ---")
print("Predicted Loan Amount: ₹", int(predicted_loan))

if approval[0] == 1:
    print("Loan Status: APPROVED")
else:
    print("Loan Status: REJECTED")

print("Approval Probability:", round(probability * 100, 2), "%")
```
